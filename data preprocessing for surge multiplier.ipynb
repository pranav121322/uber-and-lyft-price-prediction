### cheaking correlation with the surge multiplier

```python
## first of all split the data set because uber's surge multiplier is only one
df_lyft = df[df['cab_type'] == 'Lyft']
```

```python
df_lyft.info()
```

```python
from sklearn.preprocessing import LabelEncoder

lable_encoder = LabelEncoder()

df_lyft['surge_multiplier'] = lable_encoder.fit_transform(df_lyft['surge_multiplier'])

df_lyft['surge_multiplier'].unique()
```

```python
x = df_lyft[["temp" , "clouds" , "rain" , "humidity" , "wind" , "hour" , "surge_multiplier"]]
```

```python
x.info()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data["feature"] = x.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(x.values, i)
                          for i in range(len(x.columns))]
  
print(vif_data)
```
